# Função para remover os stopwords

```python
def RemoverStopWords(instancia):  # função para remover os stopwords
    instancia = instancia.lower() # transforma a frase em minusculo
    stopwords = set(nltk.corpus.stopwords.words('portuguese')) #pacote português
    palavras = [i for i in instancia.split() if not i in stopwords]  # se a palavra tiver na lista de stopwords é removida
    
    return (" ".join(palavras)) 

RemoverStopWords('CIÊNCIA DE COMPUTAÇÃO DOWNTOWN')
```

# Pré-processamento dos dados

```python
expressoes_limpeza = [
                      'formação específica' ,'formacao especifica', 'formacao especial', 'enfase',
                      'destinacao coletiva', 'complementacao estudos', 'agrupamento', 
                      'educacao superior', 'superior tecnologia', 'eixotecnologico', 'eixo tecnologico','eixotecnologicio',
                      'sequencial', 'sequncial','curso superior', 'cursos superiores', 'cursos superior', 
                      'semipresencial', 'ensino distancia','areas profissionais', 'area profissional', 'instituicoes ensino apoio', 
                      'faap', 'campus natal-central', '- anchieta', 'santa lucia', 'tangara',
                      'petropolis', 'penha','natal', 'ead', 'abc', 
                       'cursos','curso']

df = pd.read_csv("cine2000_cursos.csv")
sentenca = df["NO_CURSO"] #descrição da coluna do nome do curso da tabela original.
for line in sentenca:
        #print (line)
        new_query = RemoverStopWords(line.strip().lower())
        for expressao in expressoes_limpeza:
            new_query = new_query.replace(expressao, '').strip()
        print(new_query)  #imprimir cursos tratados        
            
with open('saida.csv', 'w') as saida:
    writer = csv.writer(saida)
    writer.writerows(new_query)
```

# Agrupamento dos cursos

```python
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
import collections

def cluster_sentences(sentences, n_clusters):  
            tfidf_vectorizer = TfidfVectorizer('') #feature extraction
        
            #contrui uma matriz para as sentenças
            tfidf_matrix = tfidf_vectorizer.fit_transform(sentences) 
            kmeans = KMeans(n_clusters=n_clusters)
            kmeans.fit(tfidf_matrix)
            clusters = collections.defaultdict(list)
            for i, label in enumerate(kmeans.labels_):
                    clusters[label].append(i)
            return dict(clusters)
        

if __name__ == "__main__":
            sentences = [ "criacao desenvolvimento sistemas web.",
                          "comercio exterior.",
                          "desenvolvimento sistemas",
                          "comercio exteriorcomercio",
                          "analise desenvolvimento sistemas",
                          "controle ambiental  meio ambiente",
                          "desenvolvimento website informatica",
                          "comercio exterior gestao negocios",
                          "controle ambiental"] 
        
            nclusters= 3 #quantidade de clusters
            
            clusters = cluster_sentences(sentences, nclusters)
            for cluster in range(nclusters):
                    print ("cluster", cluster,":")
                    for i,sentence in enumerate(clusters[cluster]):
                            print ("\tcurso ",i,": ",sentences[sentence])
```
